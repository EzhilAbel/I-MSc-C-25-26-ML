```python
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

xgb = XGBClassifier(
    n_estimators=100, learning_rate=0.1, max_depth=3,
    objective="binary:logistic", eval_metric="logloss", use_label_encoder=False
)
xgb.fit(X_clf_train, y_clf_train)
eval_clf("XGBoost", xgb)

lgbm = LGBMClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
lgbm.fit(X_clf_train, y_clf_train)
eval_clf("LightGBM", lgbm)

cat = CatBoostClassifier(
    iterations=100, learning_rate=0.1, depth=4, verbose=False, random_state=42
)
cat.fit(X_clf_train, y_clf_train)
eval_clf("CatBoost", cat)
```
